```python
data_glaxo = 'GLAXO.NS_stock_data.csv'
data_beml = 'BEML.csv'

data_glaxo = pd.read_csv(data_glaxo)
data_beml = pd.read_csv(data_beml)

print(data_glaxo.shape)
print(data_beml.shape,'\n')

print(data_glaxo.head(3),'\n')
print(data_beml.head(3))
```

```python
print(list(data_glaxo.columns))
print(list(data_beml.columns))
```

```python
glaxo_df = data_glaxo[['Unnamed: 0', 'close']]
beml_df = data_beml[['Date', 'Close']]

glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Unnamed: 0']))
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))

glaxo_df.head(5)
beml_df.head(5)
```

```python
plt.plot(glaxo_df['close'])
```

```python
plt.plot(beml_df['Close'])
```

```python
glaxo_df['gain'] = glaxo_df['close'].pct_change(periods = 1)
beml_df['gain'] = beml_df['Close'].pct_change(periods = 1)

glaxo_df.head(3)
beml_df.head(3)
```

```python
glaxo_df = glaxo_df.dropna()
beml_df = beml_df.dropna()
```

```python

plt.figure(figsize=(8,6))
plt.plot(glaxo_df.index, glaxo_df['gain'])

#plt.plot(glaxo_df['gain'])
```

```python
plt.figure(figsize = (8,6))
plt.plot(beml_df.index, beml_df['gain'])
```

```python

plt.figure(figsize=(8,6))
sns.histplot(glaxo_df['gain'], label = 'glaxo')
sns.histplot(beml_df['gain'], label = 'beml')
plt.xlabel('gain')
plt.ylabel('density')
plt.legend()
```

```python

plt.figure(figsize=(8,6))
sns.distplot(glaxo_df['gain'], label = 'glaxo')
sns.distplot(beml_df['gain'], label = 'beml')
plt.xlabel('gain')
plt.ylabel('density')
plt.legend()
```

```python

plt.figure(figsize=(8,6))
#sns.displot(glaxo_df['gain'], label = 'glaxo')
sns.displot(beml_df['gain'], label = 'beml')
plt.xlabel('gain')
plt.ylabel('density')
plt.legend()
```
