````python
#array indexing

list1 = [1,2,3,4,5]
r1 = np.array(list)

list2 = [6,7,8,9,0]
r2 = np.array([list1, list2])


#shape of the array
print(r2.shape)
print(r2)

#acessing the first row first element
print(r2[0,0])


#access the 2nd row fifth column element
print(r2[1,4])

#access the 4th and fifth element
list3 = [1,2,3,4,5,6]
r4 = np.array(list3)
print(r4[[3,5]])


#access the first 5 elements
list4 = [2,4,6,8,9,1,3]
r5 = np.array(list4)
print(r5[0:5])

r2 = np.array([list1, list2])
print(r2)
````

````python
#boolean indexing

list = [1,2,3,4,5]
r1 = np.array(list)
print(r1>2)
print(r1[r1>2])
````

````python
#slicing
list = [[1,2,3,4,5],
        [4,5,6,7,8],
        [9,8,7,6,5]]



r1 = np.array(list)
print(r1.shape)

#last 2 rows and first 2 columns (not based on index language)
print(r1[1:3, 0:3])

#first row onwards and 2nd column onwards (not based on index language)
print(r1[1:, 2:])



#2nd row and all columns (not based on index language)
b4 = r1[2:, 0:]
print(b4.shape)


#2nd row and all columns (not based on index language)
b5 = r1[2,0:]
print(b5)
print(b5.shape)
````

````python
#reshaping arrays

print(b5.shape)

#reshaping the array into one row and many rows needed by the function
b6 = b5.reshape(1,-1)
print(b6.shape)


#reshaping the array into any rows needed with only one column
b7 = b5.reshape(-1,1)
print(b7)
print(b6)
print(b5)
````
