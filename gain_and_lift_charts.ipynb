```python
import pandas as pd
import statsmodels.api as sm
```

```python
bank_df = pd.read_csv('bank.csv', sep = ';')
```

```python
bank_df.info()
```

```python
x_features = list(bank_df.columns)
x_features.remove('y')
x_features
```

```python
encoded_bank_df = pd.get_dummies(bank_df[x_features])
encoded_bank_df.head()
```

```python
y = bank_df.y.map(lambda x: int(x == 'yes'))
x = encoded_bank_df
```

```python
encoded_bank_df.info().astype(int)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x, y)
```
