```python
import pandas as pd
import numpy as np

headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

with open('car_dataset.data', 'r') as file:
    df = pd.read_csv(file, names = headers_data)



df.head()
```

```python
#replacing ? with nan

df.replace('?', np.nan, inplace = True)


df.info()
```

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df[["normalized-losses","bore","stroke","horsepower"]]= imputer_mean.fit_transform(df[["normalized-losses","bore","stroke","horsepower"]]).astype('object')
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer

imputer_mode = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df[['num-of-doors']]= imputer_mode.fit_transform(df[['num-of-doors']]).astype('object')
```

```python
df.info()
```

```python
df.dropna(subset = ['price'], axis = 0, inplace= True)

df.reset_index(drop = True, inplace = True)
```
