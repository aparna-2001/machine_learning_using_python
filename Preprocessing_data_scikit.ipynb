```python
import pandas as pd
import numpy as np

headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

with open('car_dataset.data', 'r') as file:
    df = pd.read_csv(file, names = headers_data)



df.head()
```

```python
#replacing ? with nan

df.replace('?', np.nan, inplace = True)


df.info()
```

```python
#replacing the missing values with mean

from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df[["normalized-losses","bore","stroke","horsepower","peak-rpm"]]= imputer_mean.fit_transform(df[["normalized-losses","bore","stroke","horsepower","peak-rpm"]]).astype('float')
```

```python
df.info()
```

```python
#replacing the missing value with mode

from sklearn.impute import SimpleImputer

imputer_mode = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df[['num-of-doors']]= imputer_mode.fit_transform(df[['num-of-doors']]).astype('object').astype('object')
```

```python
df.info()
```

```python
#dropping the missing values in price table

df.dropna(subset = ['price'], axis = 0, inplace= True)

df.reset_index(drop = True, inplace = True)
```

```python
df.info()
```

```python
#converting it to csv

df.to_csv('data.csv')
```

```python
df['horsepower'] = df['horsepower'].astype(float)
df['horsepower'].dtypes
```

```python
import matplotlib.pyplot as plt

plt.hist(df['horsepower'])
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower bins')
```

```python
#creating bins

bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
group_names = ['low', 'medium', 'high']

df['horsepower-binned']= pd.cut(df['horsepower'],bins, labels = group_names, include_lowest = True)

```

```python
df[['horsepower','horsepower-binned']].head(10)
```

```python
heights = df['horsepower-binned'].value_counts()
```

```python
plt.bar(group_names, heights )
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower bins')
plt.show()
```
