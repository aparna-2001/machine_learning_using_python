```python
df.replace('?',np.nan, inplace = True)
df
```

```python
df.isnull()

```

```python
missing_data = df.isnull()
for column in headers_data:
  #print(column)
  print(missing_data[column].value_counts())
  print(" ")
```

```python
#replacing the null values with mean
   #mean is replaced when there is not much outliers

avg_norm_loss = df['normalized-losses'].astype('float').mean()
#avg_norm_loss
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
#df.head(10)

avg_bore = df['bore'].astype('float').mean()
#avg_bore
df['bore'].replace(np.nan, avg_bore, inplace = True)
#df.head(10)

avg_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, avg_stroke, inplace = True)

avg_horsepow = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, avg_horsepow, inplace = True)

df.head(10)
```

```python
#replacing with mode

df['num-of-doors'].value_counts().idxmax()
df['body-style'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)

df.head(10)
```

```python
#dropping the price table

df.dropna (subset = ['price'],axis = 0,  inplace = True)
```

```python
#reseting the index after removing the rows

df.reset_index(drop = True, inplace = True)
```
